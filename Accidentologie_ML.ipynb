# Dataset Cyclistes

## Encodage des features

```python
features_cat = [ 'catv', 'obs', 'obsm', 'choc', 'manv',  'catu', 'sexe', 'trajet', 'secu1', 'jour', 'mois', 'hrmn', 'lum', 'dep', 'agg', 'int', 'atm',
             'col', 'catr', 'circ', 'prof', 'plan', 'surf', 'infra', 'situ',  'day_of_week']
features_total = features_cat + ['an_nais', 'nbv', 'vma']

df_encoded = pd.get_dummies(df_cyclistes_seuls[features_total],columns=features_cat)
X = df_encoded
y = df_cyclistes_seuls['gravity']
```

```python
standard_scaler = StandardScaler()
X_scaled = pd.DataFrame(standard_scaler.fit_transform(X), columns=X.columns)
X_scaled.head()
```

## Target en 4 catégories (indemne, blessé léger, blessé hospitalisé , tué)

### Division des données

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, random_state=36, train_size = 0.8, stratify=y)

print('Proportion des gravités dans l\'ensemble d\'entraînement : ')
display(y_train.value_counts(normalize=True))
print('Proportion des gravités dans l\'ensemble de test : ')
display(y_test.value_counts(normalize=True))
```

### Modèle LogisticRegression

```python
model_rl = LogisticRegression(random_state=36)

# Entraîner le modèle sur les données d'entraînement.
model_rl.fit(X_train, y_train)

# Faire des prédictions sur les ensembles d'entraînement et de test.
y_train_pred = model_rl.predict(X_train)
y_test_pred = model_rl.predict(X_test)

# Calculer et afficher la précision (accuracy) sur les deux ensembles.
print('Précision sur l\'ensemble d\'entraînement : ', accuracy_score(y_train, y_train_pred))
print('Précision sur l\'ensemble de test : ', accuracy_score(y_test, y_test_pred))

HTML(pd.DataFrame(confusion_matrix(y_test, y_test_pred),
             index = model_rl.classes_ + " REAL",
             columns = model_rl.classes_ + " PREDICTED").to_html())
```

```python
print(classification_report(y_test, y_test_pred))
```

### Modèle RandomForestClassifier

```python
model_rfc = RandomForestClassifier(random_state=36)

# Entraîner le modèle sur les données d'entraînement.
model_rfc.fit(X_train, y_train)

# Faire des prédictions sur les ensembles d'entraînement et de test.
y_train_pred = model_rfc.predict(X_train)
y_test_pred = model_rfc.predict(X_test)

# Calculer et afficher la précision (accuracy) sur les deux ensembles.
print('Précision sur l\'ensemble d\'entraînement : ', accuracy_score(y_train, y_train_pred))
print('Précision sur l\'ensemble de test : ', accuracy_score(y_test, y_test_pred))

HTML(pd.DataFrame(confusion_matrix(y_test, y_test_pred),
             index = model_rfc.classes_ + " REAL",
             columns = model_rfc.classes_ + " PREDICTED").to_html())
```

```python
print(classification_report(y_test, y_test_pred))
```

### Recherche des meilleurs modèles

```python
lazy = LazyClassifier(verbose=0, ignore_warnings=True, custom_metric=None)
models, predictions = lazy.fit(X_train, X_test, y_train, y_test)
```

```python
display(models)
```
